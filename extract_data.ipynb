```python
df = pd.DataFrame(pd.read_pickle('data/invoices_new.pkl'))
```

```python
with open('data/expired_invoices.txt', 'r') as file:
    expired = [int(id.strip()) for id in file.read().split(', ')]
```

```python
class DataExtractor:
    def __init__(self, df, expired):
        self.df = df
        self.expired = expired
        self.type_dict = {0: 'Material', 1: 'Equipment', 2: 'Service', 3: 'Other'}
    
    def clean(self):
        self.df = self.df.dropna().reset_index(drop=True)
        self.df['created_on'] = pd.to_datetime(self.df['created_on'], errors='coerce')
        self.df = self.df.dropna(subset=['created_on']).reset_index(drop=True)
        
    def process(self):
        rows = []
        
        for index, row in self.df.iterrows():
            if isinstance(row['items'], list):
                for item in row['items']:
                    try:
                        unit_price = float(item['item']['unit_price'])
                        quantity = int(item['quantity'])
                        item_type = item['item'].get('type', None)
                        if item_type not in self.type_dict:
                            item_type_str = 'Unknown'
                        else:
                            item_type_str = self.type_dict[item_type]

                        item_data = {
                            'invoice_id': int(row['id']),
                            'created_on': pd.to_datetime(row['created_on']),
                            'invoiceitem_id': item['item']['id'],
                            'invoiceitem_name': item['item']['name'],
                            'type': item_type_str,
                            'unit_price': unit_price,
                            'total_price': unit_price * quantity,
                            'quantity': quantity,
                            'is_expired': 1 if int(row['id']) in self.expired else 0
                        }
                        rows.append(item_data)
                    except (ValueError, TypeError):
                        continue

        new_df = pd.DataFrame(rows)
        new_df['total_price'] = pd.to_numeric(new_df['total_price'])
        invoice_totals = new_df.groupby('invoice_id')['total_price'].sum().rename('invoice_total')
        new_df = new_df.join(invoice_totals, on='invoice_id')
        new_df['percentage_in_invoice'] = new_df['total_price'] / new_df['invoice_total']
        new_df = new_df[['invoice_id', 'created_on', 'invoiceitem_id', 'invoiceitem_name', 'type', 'unit_price', 'total_price', 'percentage_in_invoice', 'is_expired']]
        
        return new_df


extractor = DataExtractor(df, expired)
clean_df = extractor.clean()
result_df = extractor.process()
result_df
```
